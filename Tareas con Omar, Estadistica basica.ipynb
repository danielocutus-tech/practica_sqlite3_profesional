### 📝 Planteamiento del Problema 1: Estadísticas de una Colección Numérica

**Instrucciones:**
Desarrolle una función en Python llamada `estadisticas()` que reciba como argumento una colección de números (lista, tupla o conjunto). La función debe procesar los datos de entrada y retornar un diccionario con las siguientes métricas estadísticas básicas:
*   Cantidad de elementos procesados (`cantidad`)
*   Suma total de los valores (`suma`)
*   Promedio aritmético (`promedio`)
*   Valor mínimo encontrado (`minimo`)
*   Valor máximo encontrado (`maximo`)

**Restricciones y Reglas de Negocio:**
1.  **Validación de Vacíos:** Si la colección recibida está vacía, la función no debe continuar con el cálculo matemático; en su lugar, debe lanzar explícitamente una excepción de tipo `ValueError` con un mensaje descriptivo.
2.  **Robustez y Tipado:** El sistema debe validar que los elementos internos sean estrictamente numéricos (`int` o `float`) y estar protegido contra valores indeterminados como `NaN` (*Not a Number*). En caso de detectar una anomalía, debe lanzar un `TypeError`.
3.  **Control de Flujo:** En su celda de pruebas, implemente bloques `try/except` para interceptar de manera limpia estas excepciones y mostrar un mensaje amigable al usuario en lugar de romper la ejecución del cuaderno.

````python
import math

def estadisticas(numeros):
    
    # 1. ¿Es una estructura válida?
    if not isinstance(numeros, (list, tuple, set)):
        raise TypeError("El argumento debe ser una colección de números.")
    
    # Estandariza el parámetro a tipo lista para evaluar sus elementos uno a uno
    numeros = list(numeros)
        
    # 2. ¿Está vacía? Condición requerida por la instrucción del Profesor Omar Peña; lanzar una excepción (ValueError).
    if not numeros:
        raise ValueError("La lista no puede estar vacía.")
        
    # 3. ¿Todos son números válidos?
    for x in numeros:
        if not isinstance(x, (int, float)) or (isinstance(x, float) and math.isnan(x)):
            raise TypeError("Todos los elementos deben ser números válidos.")

    # Procesamiento de los números en estadística
    minimo = min(numeros)
    maximo = max(numeros)
    promedio = sum(numeros) / len(numeros)
    suma = sum(numeros)
    cantidad = len(numeros)
    
    return {
        "cantidad": cantidad, 
        "suma": suma, 
        "promedio": promedio, 
        "minimo": minimo, 
        "maximo": maximo
    }
````

### 🛠️ Cláusulas de Guarda (Manejo de Errores Previo)

#### 1. Validación de la Estructura de Datos
* **Código:** `if not isinstance(numeros, (list, tuple, set)):`
* **Excepción lanzada:** `TypeError("El argumento debe ser una colección de números.")`
* **¿Qué hace?** Filtra y descarta datos individuales (como un número entero suelto `5`), cadenas de texto (`"hola"`) o valores nulos del sistema (`None`). Asegura que la función reciba obligatoriamente una colección iterable que contenga elementos en su interior.

#### 2. Validación de Colección Vacía
* **Código:** `if not numeros:`
* **Excepción lanzada:** `ValueError("La lista no puede estar vacía.")`
* **¿Qué hace?** Detiene el programa si la lista, tupla o conjunto no contiene ningún elemento. Esto previene errores críticos del sistema, como intentar calcular el promedio de la nada, lo cual provocaría una división entre cero (`ZeroDivisionError`).

#### 3. Validación de Tipo de Elemento (Datos Válidos)
* **Código:** `for x in numeros: if not isinstance(x, (int, float)) or ...`
* **Excepción lanzada:** `TypeError("Todos los elementos deben ser números válidos.")`
* **¿Qué hace?** Analiza uno a uno los elementos internos de la colección. Si detecta que algún elemento no es un número (por ejemplo, si hay un texto como `[1, 2, "tres"]`) o si encuentra un valor indefinido matemático como `NaN` (*Not a Number*), interrumpe el proceso para evitar fallas en funciones nativas de Python como `sum()`, `min()`, o `max()`.

### 💡 'Problema 2. Sistema de calificaciones'.
### Optimización Avanzada del Sistema ( solución 2/2)

Para resolver este problema cumpliendo con los objetivos de modularidad (uso de **funciones**) y optimización de código nativo de Python (**comprensión de diccionarios y listas**), se estructuró una solución compacta y eficiente:

#### 1. Estructura Principal (`def`)
* Todo el proceso se empaqueta en la función `sistema_calificaciones(alumnos)`. Esto permite reutilizar el código con cualquier otro grupo de estudiantes en el futuro sin tener que reescribir las celdas.

#### 2. Comprensión de Diccionarios (Generación de Promedios)
* **Código:** `promedios = {alumno: round(sum(notas) / len(notas), 1) if notas else 0.0 for alumno, notas in alumnos.items()}`
* **¿Qué hace?** Sustituye el bucle tradicional de 6 líneas por una sola línea elegante. Itera por cada alumno, suma sus notas, calcula el promedio, lo redondea a un decimal y maneja de forma segura el caso de que un alumno no tenga calificaciones registradas.

#### 3. Comprensión de Listas (Filtro de Reprobados)
* **Código:** `reprobados = [alumno for alumno, prom in promedios.items() if prom < 70]`
* **¿Qué hace?** Es el núcleo de lo solicitado en el objetivo. Recorre el nuevo diccionario de promedios y va guardando en una lista únicamente los nombres de los estudiantes cuya nota sea estrictamente menor a 70.

````python

def sistema_calificaciones(alumnos):
    # 1. Generar el diccionario de promedios usando Comprensión de Diccionarios
    promedios = {
        alumno: round(sum(notas) / len(notas), 1) if notas else 0.0 
        for alumno, notas in alumnos.items()
    }
    
    # 2. Determinar el promedio más alto
    if promedios:
        alumno_mas_alto = max(promedios, key=promedios.get)
        promedio_mas_alto = promedios[alumno_mas_alto]
    else:
        alumno_mas_alto, promedio_mas_alto = None, 0.0
        
    # 3. Filtrar alumnos reprobados usando Comprensión de Listas
    reprobados = [alumno for alumno, prom in promedios.items() if prom < 70]
    
   
    
    # Mostrar resultados en pantalla de forma limpia
    print("--- 📊 RESULTADOS DEL SISTEMA ---")
    print(f"El alumno con el promedio más alto es: {alumno_mas_alto} ({promedio_mas_alto})")
    
    if reprobados:
        print(f"Alumnos reprobados (< 70): {', '.join(reprobados)}")
    else:
        print("¡Ningún alumno reprobó!")
        
    # Retornar el nuevo diccionario generado tal como pide la salida esperada
    return promedios

# --- PRUEBA CON TUS DATOS DEL NOTEBOOK ---
alumnos_grupo = { 
    "Ana": [90, 80, 95], 
    "Luis": [70, 88, 91], 
    "Carlos": [100, 95, 98], 
    "María": [60, 75, 80] 
}

# Ejecución de la función
diccionario_final = sistema_calificaciones(alumnos_grupo)
print("\nSalida del nuevo diccionario:")
display(diccionario_final)
````

### 🔧 Métodos y Módulos Nativos de Python utilizados

Para lograr un código eficiente y avanzado en el **Problema 2**, aprovechamos las herramientas que Python incluye por defecto en su núcleo, lo que significa que **no requieren la descarga ni instalación de librerías externas**:

#### 1. Módulo Nativo (o Librería Estándar)

#### 2. Métodos Nativos de los Diccionarios
Los métodos son "habilidades" o funciones que pertenecen exclusivamente a un tipo de objeto. En este sistema usamos dos muy importantes:

* **`.items()`**: 
  * *¿Para qué sirve?* Funciona como un desempacador. Por defecto, un bucle `for` solo lee los nombres de los alumnos (las claves). Al usar `.items()`, le indicamos a Python que nos entregue la pareja completa: el nombre (`alumno`) y sus calificaciones (`notas` o `prom`) al mismo tiempo.
* **`.get`**: 
  * *¿Para qué sirve?* Sirve para consultar el valor de una clave. Lo usamos como un "asistente" dentro de la función `max(promedios, key=promedios.get)`. Esto obliga a Python a comparar los promedios numéricos en lugar de los nombres en orden alfabético, devolviendo correctamente al estudiante con la calificación más alta.

## 💡Solución del "Sistema de Calificaciones" (solución 3/3)

### Utilizando nuestra función **```Estadisticas()```**

````python
def sistema_calificaciones_con_estadisticas(alumnos):
    # 1. Generar el diccionario de promedios reutilizando estadisticas()
    # Usamos un bloque try/except interno para manejar de forma segura si un alumno no tiene notas
    promedios = {}
    for alumno, notas in alumnos.items():
        try:
            # Reutilizamos la función del Problema 1; ella calcula todo y nos da el promedio
            resultado_estadistico = estadisticas(notas)
            promedios[alumno] = round(resultado_estadistico["promedio"], 1)
        except ValueError:
            # Si estadisticas() lanza ValueError (lista vacía), le asignamos 0.0
            promedios[alumno] = 0.0
            
    # 2. Determinar el promedio más alto (se mantiene igual)
    if promedios:
        alumno_mas_alto = max(promedios, key=promedios.get)
        promedio_mas_alto = promedios[alumno_mas_alto]
    else:
        alumno_mas_alto, promedio_mas_alto = None, 0.0
        
    # 3. Filtrar alumnos reprobados usando Comprensión de Listas (se mantiene igual)
    reprobados = [alumno for alumno, prom in promedios.items() if prom < 70]
    
    # Mostrar resultados en pantalla
    print("--- 📊 RESULTADOS DEL SISTEMA (USANDO ESTADISTICAS()) ---")
    print(f"El alumno con el promedio más alto es: {alumno_mas_alto} ({promedio_mas_alto})")
    if reprobados:
        print(f"Alumnos reprobados (< 70): {', '. join(reprobados)}")
    else:
        print("¡Ningún alumno reprobó!")
        
    return promedios

# --- EJECUCIÓN CON TUS DATOS ---
alumnos_grupo = {
    "Ana": [90, 80, 95],
    "Luis": [70, 88, 91],
    "Carlos": [100, 95, 98],
    "María": [60, 75, 80]
}

diccionario_final = sistema_calificaciones_con_estadisticas(alumnos_grupo)
print("\nSalida del nuevo diccionario:")
display(diccionario_final)
````
